# **Setup**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
```

# **Import datasets**

```python
flights=sn.load_dataset("flights")
flights.head()
```

```python
flights.shape
```

```python
flights_feb=flights[flights['month']=='Feb']
flights_feb

```

# **Line plot**

```python
sn.set_style('darkgrid')
sn.lineplot(data=flights, x="year", y="passengers",style='month',hue="month")
```

```python
sn.lineplot(data=flights_feb, x="year", y="passengers")
```

```python
fmri=sn.load_dataset('fmri')
fmri
```
