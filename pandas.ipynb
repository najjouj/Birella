# **Pivot Table**

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic['sex'].value_counts()
```

```python
titanic.pivot_table(index='who', columns='class', values='age', aggfunc='mean')
```

```python
titanic.pivot_table(index='who', columns='class', values='survived', aggfunc='mean')
```

```python
titanic.pivot_table(index='who', values='survived', aggfunc='mean')
```

```python
titanic.pivot_table(index='sex', columns='class', values='survived', aggfunc='mean')
```

```python
titanic.pivot_table(index='sex', columns='class', values='survived', aggfunc='count')
```

```python
titanic.pivot_table(index='sex', columns='class', values=['survived','age'], aggfunc='count')
```

```python
titanic.pivot_table(index='sex', columns='class', values=['survived','age'], aggfunc={'survived':'mean','age':'mean'})
```

```python
titanic.pivot_table(index='who', columns='class', values=['survived','age'], aggfunc={'survived':'median','age':'mean'})
```

```python
titanic.pivot_table(index=['who','sex'], columns='class', values=['survived','age'], aggfunc={'survived':'median','age':'mean'})
```
